# Preface
We are agents in an age-old battle between Dynamic and Static systems, and I stand firmly with the former. In clinical and public health literature, typical graphical outputs—like the static Kaplan-Meier survival curves—provide snapshots of population-level survival probabilities over time. These are orthodox, comforting in their finality and simplicity. They are, however, incomplete. Static representations are rigid and fail to capture the personalized, interactive nature of risk—something dynamic systems can do with precision.

In this Jupyter Book, I offer more than mere orthodoxy. I introduce a dynamic, interactive web application that personalizes risk for individuals based on 58-100 demographic and clinical characteristics. Hosted on GitHub, this app uses .js and .html to take static orthodoxy and explode it with possibilities, allowing for a real-time exploration of personal data. The output is dynamic: fluid, interactive, and adaptive. It reflects the complexity of human experience rather than forcing it into predetermined subgroups or categories.
